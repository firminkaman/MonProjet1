**Partie 1: Importation des bibliothèques nécessaires**


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

 **Partie 2: Génération de données
Générez un ensemble de données en utilisant la fonction make_blobs de scikit-learn.**








```python
# Génération de données
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
```

 **Partie 3: Visualisation des données
Tracez le nuage de points pour visualiser les données générées.**

```python
# Visualisation des données
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title('Données Générées')
plt.show()
```

 **Partie 4: Application de l'algorithme K-Means
Appliquez l'algorithme K-Means sur les données.**




```python
# Initialisation et ajustement du modèle K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
```

 **Partie 5: Visualisation des clusters
Visualisez les clusters créés par l'algorithme K-Means.**




```python
# Visualisation des clusters
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Clusters avec les centres')
plt.show()
```

**Questions Guidées :**

- Que représentent les centres rouges dans le graphique ?

Les centres rouges dans le graphique représentent les centres des clusters calculés par l'algorithme K-Means. L'algorithme K-Means cherche à partitionner les données en un certain nombre de clusters (dans ce cas, 4 clusters, comme spécifié par n_clusters=4). Une fois que les clusters sont formés, les centres de ces clusters sont déterminés, et ils sont affichés dans le graphique en tant que points rouges.

- Quelle est la différence entre les données générées et les clusters obtenus après l'application de K-Means ?

Les "données générées" sont les points dispersés dans l'espace bidimensionnel, créés synthétiquement à l'aide de la fonction make_blobs. Les "clusters obtenus après l'application de K-Means" sont les groupes identifiés par l'algorithme K-Means lorsqu'il a été appliqué à ces données. Les centres rouges dans le deuxième graphique représentent les centres de ces clusters identifiés. En d'autres termes, les données générées sont le point de départ, tandis que les clusters résultent de l'application de l'algorithme K-Means pour regrouper ces données.

- Modifiez le code pour générer un ensemble de données avec 5 centres au lieu de 4. Comment cela affecte-t-il les résultats de K-Means ?





```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Génération de données avec 5 centres
X, y = make_blobs(n_samples=300, centers=5, random_state=42)

# Visualisation des données générées
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title('Données Générées')
plt.show()
```

```python
# Initialisation et ajustement du modèle K-Means avec 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
```

```python
# Visualisation des clusters avec les centres
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Clusters avec les centres')
plt.show()
```
